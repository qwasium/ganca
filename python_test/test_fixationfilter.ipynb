### aka

````python
aka_timeseries = pd.read_csv('../../test_data/aka_timeseries.csv')

# change int to bool
aka_timeseries['validity_gaze_left'] = aka_timeseries['validity_gaze_left'].astype('bool')
aka_timeseries['validity_gaze_right'] = aka_timeseries['validity_gaze_right'].astype('bool')

# create vector gaze_point-gaze_origin
aka_timeseries['gaze_vector_left_x'] = aka_timeseries['gaze_point_ucs_left_x'] - aka_timeseries['gaze_origin_ucs_left_x']
aka_timeseries['gaze_vector_left_y'] = aka_timeseries['gaze_point_ucs_left_y'] - aka_timeseries['gaze_origin_ucs_left_y']
aka_timeseries['gaze_vector_left_z'] = aka_timeseries['gaze_point_ucs_left_z'] - aka_timeseries['gaze_origin_ucs_left_z']
aka_timeseries['gaze_vector_right_x'] = aka_timeseries['gaze_point_ucs_right_x'] - aka_timeseries['gaze_origin_ucs_right_x']
aka_timeseries['gaze_vector_right_y'] = aka_timeseries['gaze_point_ucs_right_y'] - aka_timeseries['gaze_origin_ucs_right_y']
aka_timeseries['gaze_vector_right_z'] = aka_timeseries['gaze_point_ucs_right_z'] - aka_timeseries['gaze_origin_ucs_right_z']
aka_timeseries
````

````python
aka_left = pyFixationFilter.FixationFilter(
	timestampInInit=aka_timeseries['eyetracker_timestamp'].to_list(),
	validityInInit=aka_timeseries['validity_gaze_left'].to_list(),
	gazeVectorXInit=aka_timeseries['gaze_vector_left_x'].to_list(),
	gazeVectorYInit=aka_timeseries['gaze_vector_left_y'].to_list(),
	gazeVectorZInit=aka_timeseries['gaze_vector_left_z'].to_list()
)
aka_left_ivt = aka_left.IVTFilter(velocityThresholdIn=30)

# generator object
print(list(aka_left.isFixation))
print(list(aka_left.validityOut))
print(list(aka_left.timestampOut))
print(list(aka_left.velocity))
````

````python
aka_right = pyFixationFilter.FixationFilter(
	timestampInInit=aka_timeseries['eyetracker_timestamp'].to_list(),
	validityInInit=aka_timeseries['validity_gaze_right'].to_list(),
	gazeVectorXInit=aka_timeseries['gaze_vector_right_x'].to_list(),
	gazeVectorYInit=aka_timeseries['gaze_vector_right_y'].to_list(),
	gazeVectorZInit=aka_timeseries['gaze_vector_right_z'].to_list()
)
aka_right_ivt = aka_right.IVTFilter(velocityThresholdIn=30)

# generator object
print(list(aka_right.isFixation))
print(list(aka_right.validityOut))
print(list(aka_right.timestampOut))
````

````python
aka_left_python = pyTestFixationFilter(
	time_stamp_in=aka_timeseries['eyetracker_timestamp'].to_list(),
	validity_in=aka_timeseries['validity_gaze_left'].to_list(),
	gaze_vector_x=aka_timeseries['gaze_vector_left_x'].to_list(),
	gaze_vector_y=aka_timeseries['gaze_vector_left_y'].to_list(),
	gaze_vector_z=aka_timeseries['gaze_vector_left_z'].to_list(),
	)
aka_left_python.IVT_filter(velocity_threshold=30)

print(aka_left_python.is_fixation)
print(aka_left_python.validity_out)
print(aka_left_python.time_stamp_out)

print(pyTestFunc.test_vector(aka_left.isFixation, aka_left_python.is_fixation))
````

````python
aka_right_python = pyTestFixationFilter(
	time_stamp_in=aka_timeseries['eyetracker_timestamp'].to_list(),
	validity_in=aka_timeseries['validity_gaze_right'].to_list(),
	gaze_vector_x=aka_timeseries['gaze_vector_right_x'].to_list(),
	gaze_vector_y=aka_timeseries['gaze_vector_right_y'].to_list(),
	gaze_vector_z=aka_timeseries['gaze_vector_right_z'].to_list(),
	)
aka_right_python.IVT_filter(velocity_threshold=30)

print(aka_right_python.is_fixation)
print(aka_right_python.validity_out)
print(aka_right_python.time_stamp_out)

print(len(aka_right.isFixation))

print(pyTestFunc.test_vector(aka_right.isFixation, aka_right_python.is_fixation))
````
